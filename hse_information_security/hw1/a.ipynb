```python
def extended_gcd(a, b):
    if a == 0:
        return b, 0, 1
    gcd, x1, y1 = extended_gcd(b % a, a)
    x = y1 - (b // a) * x1
    y = x1
    return gcd, x, y

def mod_inverse(a, n):
    gcd, x, y = extended_gcd(a, n)
    if gcd != 1:
        raise ValueError(f"Обратного элемента не существует для {a} по модулю {n}")
    else:
        return x % n

def solve_for_x(a, b, y, n):
    n_squared = n ** 2
    
    a_inv = mod_inverse(a, n_squared)
    
    x = a_inv * (y - b) % n_squared
    
    return x
```

```python
from math import gcd

n = 26

def decrypt(bigrams, a, b):
    decrypted_text = ''
    for y in bigrams:
        x = solve_for_x(a, b, y, n)
        decrypted_text += chr(x // n + ord('a')) + chr(x % n + ord('a'))
    print(a, b, decrypted_text)

def f(text):
    bigrams = list(map(int, text.split()))
    
    for a in range(n):
        if gcd(a, n) != 1:
            continue
        for b in range(n):
            decrypt(bigrams, a, b)


```
